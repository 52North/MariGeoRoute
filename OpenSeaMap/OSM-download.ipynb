## Checking data download from geofabrik portal using url request

```python
import requests
```

```python
def download_file(url):
    local_filename = url.split('/')[-1]
    # NOTE the stream=True parameter below
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192): 
                # If you have chunk encoded response uncomment if
                # and set chunk_size parameter to None.
                #if chunk: 
                f.write(chunk)
    return local_filename
```

```python
#Retrieving the metadata from the geofrabik portal
r=requests.get('https://download.geofabrik.de/index-v1-nogeom.json') # this is metadata without boundary geometry
r.text
```

```python
#loading metadata in json format
import json
js=json.loads(r.text)
js
```

```python
a=js['features']
```

```python
#All existing dataset urls from geofabrik portal in pbf format
for i in a:
    print(i['properties']['urls']['pbf'])
```

```python
# Creating the list of the dataset url for european countries

a=js['features'][:]
url_list = []
for i in a:
# Selecting european countries and Filtering non-administrative dataset url from the url list
    if (i['properties']['urls']['pbf'].split('/')[3]=='europe' and 
        len(i['properties']['urls']['pbf'].split('/'))==5 and i['properties']['id'] != 'dach'
        and i['properties']['id'] != 'alps') and i['properties']['id'] != 'britain-and-ireland' :
        #print(i['properties']['urls']['pbf'])
        url_list.append(i['properties']['urls']['pbf'])
print(url_list)
```

```python
# Downloading all countries dataset one by one 
# This needs to be checked
def download_file(urls):
    for url in urls:
        local_filename = url.split('/')[-1]
        # NOTE the stream=True parameter below
        with requests.get(url, stream=True) as r:
            r.raise_for_status()
            with open(local_filename, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192): 
                    # If you have chunk encoded response uncomment if
                    # and set chunk_size parameter to None.
                    #if chunk: 
                    f.write(chunk)
    return local_filename
```

```python
#Downloads osm data for all european countries
download_file(url_list)
```
